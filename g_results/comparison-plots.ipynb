```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

percentages = np.array([75, 60, 45, 30, 15])

# Model names
models = ["GNN", "ChemBERTa", "QR Code", "Circular FP"]

# MSE values
mse_values = [
    [22.9795, 45.2777, 33.4616, 58.0983, 105.5618],  # GNN
    [22.6927, 35.7943, 39.5323, 58.8302, 71.9250],  # ChemBERTa
    [49.9385, 61.5937, 80.0927, 127.5616, 256.1001],  # QR Code 
    [287.5763, 307.7243, 360.1455, 434.4015, 571.5742],  # Circular FP
]

# MAE values
mae_values = [
    [2.7877, 3.3917, 2.9739, 4.7256, 6.6468],
    [3.4568, 4.4287, 4.4157, 5.8742, 6.4864],
    [3.4024, 4.1052, 4.6064, 6.1918, 10.5020],
    [11.9760, 12.5052, 13.6055, 15.3522, 17.8153],
]

# R^2 values
r2_values = [
    [0.9854, 0.9708, 0.9787, 0.9633, 0.9340],
    [0.9858, 0.9773, 0.9755, 0.9635, 0.9554],
    [0.9692, 0.9610, 0.9501, 0.9208, 0.8404],
    [0.8199, 0.8052, 0.7772, 0.7304, 0.6454],
]

# Inference times
inference_times = [
    [8.3334],
    [0.6400],
    [10.5091],
    [0.7852],
]

# Plotting function
def plot_metric(metric_values, title, ylabel):
    plt.figure(figsize=(8, 5))
    for i, model in enumerate(models):
        plt.plot(percentages, metric_values[i], marker='o', linestyle='-', label=model)
    plt.xlabel("Training Data Percentage")
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# Generate plots
plot_metric(mse_values, "Mean Squared Error (MSE) Comparison", "MSE")
plot_metric(mae_values, "Mean Absolute Error (MAE) Comparison", "MAE")
plot_metric(r2_values, "R-Squared (R²) Comparison", "R²")

inference_times_flat = [time[0] for time in inference_times]  # Extract single values from nested lists

plt.figure(figsize=(8, 5))
sns.barplot(x=inference_times_flat, y=models, orient='h', palette='viridis')
plt.xlabel("Inference Time (s) per 1000 samples")
plt.title("Inference Time Comparison")
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```
